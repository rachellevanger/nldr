# Locally-linear Embedding (LLE)

Given $N$ points in $D$-dimensional space, this algorithm will embed the points into a $d$-dimensional space, preserving the $K$ nearest neighbor relations. Code modeled after the MATLAB code from https://www.cs.nyu.edu/~roweis/lle/code.html.

### Generate data set and set variables.

```python
import math as m
import numpy as np

# LLE Parameters
N = 2000 # Number of points
K = 12   # Number of neighbors
d = 2    # Number of embedding dimensions

# Swiss roll example.
tt = (3*m.pi/2)*(1+2*np.random.rand(1,N)) # Random points
height = 21*np.random.rand(1,N) # Random heights

X = np.array([np.multiply(tt, np.cos(tt)), height, np.multiply(tt,np.sin(tt))])
color = tt*np.ones(np.shape(height)[1])
```

### View the data set in 3D.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

%pylab inline

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[0,0,:], X[1,0,:], X[2,0,:], c=color, marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### Perform LLE.

##### Find the K nearest neighbors of each point.

```python
import scipy.spatial.distance as sdist

# Generate distance matrix
X_dist = sdist.squareform(sdist.pdist(np.transpose(X[:,0,:]), 'Euclidean'))

# Compute the K nearest neighbors, one column for each data point
neighbors = np.zeros((K, N))
for i in range(0,N):
    neighbors[:,i] = np.argsort(X_dist[:,i])[1:(K+1)]
neighbors = neighbors.astype(int)
```

##### Solve for the reconstruction weights.

```python
# If K > D=initial dimension of space, then need to regularize.
D = np.shape(X)[0]

reg = 0
if K > D:
    reg = 0.001

# Create weight matrix
W = np.zeros((K,N))
for i in range(0,N):
    Z = X[:,0,neighbors[:,i]] - np.transpose(np.tile(X[:,0,i],(K, 1)))
    C = np.dot(np.transpose(Z),Z)
    C = C + np.identity(K)*reg*np.trace(C)
    W[:,i] = np.linalg.solve(C, np.ones(K))
    W[:,i] = W[:,i]/sum(W[:,i])
```

##### Compute the embedding.

```python
# Compute matrix that embedding relies on
M = np.identity(N)
for i in range(0,N):
    w = np.reshape(W[:,i],(1,np.shape(W[:,i])[0]))
    j = neighbors[:,i].tolist()
    for jj in range(0,len(j)):
        M[i,j[jj]] = M[i,j[jj]] - w[0,jj]
        M[j[jj],i] = M[j[jj],i] - w[0,jj]
        M[j[jj],j] = M[j[jj],j] + np.multiply(np.transpose(w),w)[jj,:]

# Construct the embedding matrix
evals, Y = np.linalg.eigh(M)
Y = np.transpose(Y[:,1:(d+1)])*m.sqrt(N)
```

### Plot the results.

```python
%pylab inline

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(Y[0,:], Y[1,:], c=color, marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
