# Principal Component Analysis (PCA)

Given a collection of datapoints described as vectors in some space:

1. Compute the covariance matrix of the dataset.
2. Diagonalize the covariance matrix to get an orthogonal set of eigenvectors.

The eigenvectors give the new basis along which the data is organized, and the magnitude of the eigenvalues give the degree to which the data is organized along each eigenvector.

### Generate the data set

```python
import numpy as np

# Dataset is a collection of points arranged in (roughly) a line through 3-space, plus some noise.

# Set up dataset parameters
N = np.array([2, 1, 3]) # Vector along which data is organized
t = np.random.rand(50,1) # Random times in [0,1) that give the number and spacing of points
scale = 10 # Factor with which to stretch out the data points
noise = 3 # Noise scale

# Generate the data set
t_scaled = t*10
X = np.zeros((np.shape(t)[0],np.shape(N)[0])) # Set up matrix to hold points; rows are points, columns are coordinates
X_noise = np.random.rand(np.shape(X)[0], np.shape(X)[1])*noise # Noise matrix
for i in range(0,np.shape(N)[0]):
    X[:,i] = np.transpose(N[i]*t_scaled)
X = X + X_noise
```

### Display the data set before running PCA

```python
# Plot the first three coordinates of the data set to convince viewers the data is in a line

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

%pylab inline

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], c='r', marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### Perform PCA

```python
from scipy import linalg as LA

# Compute the covariance matrix of the data set
C = np.cov(X, rowvar=0)

# Compute the eigenvectors and eigenvalues of the covariance matrix
evals, evecs = LA.eigh(C)

# Sort the eigenvalues in decreasing order
idx = np.argsort(evals)[::-1]
evecs = evecs[:,idx]
# Sort the eigenvectors in this same corresponding order
evals = evals[idx]
```

### Analyze PCA results

Here, we look at the eigenvalues returned by the PCA analysis to determine how many dimensions along which the data is organized.

```python
# Show eigenvalues
evals
```

The first entry in the vector should be large compared to the others in the data set. This indicates that the data set is organized primarily in a single direction.

### Plot data set according to (sorted) basis for PCA

```python
# Mean-center the data
X -= X.mean(axis=0)

# Perform transformation via eigenvectors
Y = np.dot(X, evecs)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(Y[:,0], Y[:,1], Y[:,2], c='r', marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
